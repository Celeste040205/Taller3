Encuentre el polinomio de Lagrange para los siguientes datos y grafique:
***(âˆ’2,5),(1,7),(3,11),(7,34)***

```python
def f(x):
    return (43/1080)*x**3 + (101/540)*x**2 + (793/1080)*x + (1087/180)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
valores = np.linspace(-10, 10, 600)
fx=[]

for numero in valores:
    fx.append(f(numero))

plt.figure(figsize=(10,6))
plt.plot(valores, fx, label='Lagrange')
x_points = np.array([-2, 1, 3, 7])
y_points = np.array([5, 7, 11, 34])
plt.scatter(x_points, y_points, color='red', label='Puntos dados', zorder=5)
plt.xlim(-5,10)
plt.ylim(-5,40)
plt.xlabel('x')
plt.ylabel('Valor de la funciÃ³n')
plt.title('GrÃ¡ficas de polinomio de Lagrange')
plt.legend()
plt.grid(True)
plt.show()
```
