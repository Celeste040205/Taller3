Grafique las curvas de las series de Taylor de varios Ã³rdenes para los siguientes casos:

$ln(x), \quad x_0=1$

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return math.log(x)

def f_1(x):
    return x-1

def f_2(x):
    return x-1-(1/2)*(x-1)**2
    
def f_3(x):
    return x-1-(1/2)*(x-1)**2+(1/3)*(x-1)**3
    
def f_4(x):
    return x-1-(1/2)*(x-1)**2+(1/3)*(x-1)**3-(1/4)*(x-1)**4
```

```python
valores = np.linspace(0.1, 10, 500)
fx=[]
fx2=[]
fx3=[]
fx4=[]
fx1=[]

for numero in valores:
    fx.append(f(numero))
    fx1.append(f_1(numero))
    fx2.append(f_2(numero))
    fx3.append(f_3(numero))
    fx4.append(f_4(numero))
```

```python
plt.figure(figsize=(10,6))
plt.plot(valores, fx, label=r"$\ln(x)$")
plt.plot(valores, fx1, label=r"$x - 1$",linestyle='--')
plt.plot(valores, fx2, label=r"$x - 1 - \frac{1}{2}(x - 1)^2$",linestyle='--')
plt.plot(valores, fx3, label=r"$x - 1 - \frac{1}{2}(x - 1)^2 + \frac{1}{3}(x - 1)^3$",linestyle='--')
plt.plot(valores, fx4, label=r"$x - 1 - \frac{1}{2}(x - 1)^2 + \frac{1}{3}(x - 1)^3 - \frac{1}{4}(x - 1)^4$",linestyle='--')

plt.xlim(0,4)
plt.ylim(-2.5, 2.5)
plt.xlabel('x')
plt.ylabel('Valor de la funciÃ³n')
plt.title('GrÃ¡ficas de f(x) y sus derivadas')
plt.legend()
plt.grid(True)
plt.show()
```
