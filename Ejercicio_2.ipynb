```python
import numpy as np
def f(x):
    return 1
def f1(x):
    return 1 + (x)**2
def f2(x):
    return 1 + (x)**2 + (x)**3
def f3(x):
    return 1 + (x)**2 + (x)**3 + (x)**4
def f4(x):
    return 1 + (x)**2 + (x)**3 + (x)**4 + (x)**5
```
