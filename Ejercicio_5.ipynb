Encuentre el polinomio de Lagrange para los siguientes datos y grafique:
***(1,1),(2,2),(3,2)***

```python
def f(x):
    return (1/2)*(x**2-5*x+6)-2*x**2+8*x-6+x**2-3*x+2
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
valores = np.linspace(-10, 10, 500)
fx=[]

for numero in valores:
    fx.append(f(numero))

plt.figure(figsize=(10,6))
plt.plot(valores, fx, label='Lagrange')
plt.plot(1, 1, 'ro', label="Punto (1, 1)")
plt.plot(2, 2, 'ro', label="Punto (2, 2)")
plt.plot(3, 2, 'ro', label="Punto (3, 2)")
plt.xlim(0,4)
plt.ylim(-2.5, 2.5)
plt.xlabel('x')
plt.ylabel('Valor de la función')
plt.title('Gráficas de polinomio de Lagrange')
plt.legend()
plt.grid(True)
plt.show()
```
